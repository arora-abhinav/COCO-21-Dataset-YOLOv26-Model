```python
from ultralytics import YOLO
```

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name())
```

```python
model = YOLO("yolo26x-pose.pt") 

results = model.train(
    data="custom_coco.yaml",
    epochs=50,
    imgsz=640,
    batch=2,          # Safe physical batch size for 2060 RTX
    optimizer="MuSGD",
    device=0,
    workers=0,
    cache="disk"
)

model.save("coco21.pt")
```

```python
model = YOLO("best.pt")
metrics = model.val(data="custom_coco.yaml", device=0)
```
